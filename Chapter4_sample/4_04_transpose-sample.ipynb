▲리스트4.15 : 전치한 후 행렬곱을 실시한다

```python
import numpy as np

a = np.array([[0, 1, 2],
             [1, 2, 3]])  # 2×3의 행렬
b = np.array([[0, 1, 2],
             [1, 2, 3]])  # 2×3의 행렬 이대로는 행렬곱을 할 수 없다

# print(np.dot(a, b))  # 전치하지 않고 행렬곱을 취하면 에러
print(np.dot(a, b.T))  # 전치에 의해 행렬곱이 가능으로
```
