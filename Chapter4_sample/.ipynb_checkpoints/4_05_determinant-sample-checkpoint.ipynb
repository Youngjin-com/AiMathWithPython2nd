▲리스트4.18：`eye()`함수를 사용해서 단위행렬을 만든다

```python
import numpy as np

print(np.eye(2))  # 2×2의 단위 행렬
print()
print(np.eye(3))  # 3×3의 단위 행렬
print()
print(np.eye(4))  # 4×4의 단위 행렬
```

▲리스트4.19 : `linalg.det()`함수를 사용해서 행렬식을 구한다

```python
import numpy as np

a = np.array([[1, 2],
              [3, 4]])
print(np.linalg.det(a))  # 행렬식이 0이 되지 않는 경우 

b = np.array([[1, 2],
              [0, 0]])
print(np.linalg.det(b))  # 행렬식이 0이 되는 경우
```
