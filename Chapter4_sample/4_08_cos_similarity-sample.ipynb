▲리스트4.31 : 코사인 유사도를 계산한다

```python
import numpy as np

def cos_sim(vec_1, vec_2):
    return np.dot(vec_1, vec_2) / (np.linalg.norm(vec_1) * np.linalg.norm(vec_2))  

a = np.array([2, 2, 2, 2])
b = np.array([1, 1, 1, 1])  # a와 같은 방향
c = np.array([-1, -1, -1, -1])  # a와 반대 방향

print("--- a와 b의 코사인 유사도 ----")
print(cos_sim(a, b))

print("--- a와 c의 코사인 유사도 ----")
print(cos_sim(a, c))
```
