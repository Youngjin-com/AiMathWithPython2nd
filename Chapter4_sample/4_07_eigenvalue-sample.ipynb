▲리스트4.28 : `linalg.eig()`함수를 사용해 고유값과 고유 벡터를 구한다

```python
import numpy as np

a = np.array([[3, 1],
              [2, 4]])

ev = np.linalg.eig(a)  # 고유값과 고유 벡터를 동시에 구한다

print(ev[0])  # 첫 요소는 고유값 

print()

print(ev[1])  # 다음 요소는 고유 벡터
```
