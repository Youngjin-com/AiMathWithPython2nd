▲리스트3.19 : NumPy의 sum함수로 총합을 구한다

```python
import numpy as np

a = np.array([1, 3, 2, 5, 4])  # a1부터 a5까지
y = np.sum(a)  # 총합
print(y)
```

▲리스트3.20 : NumPy의 prod함수로 총곱을 구한다

```python
import numpy as np

a = np.array([1, 3, 2, 5, 4])  # a1부터 a5까지
y = np.prod(a)  # 총곱
print(y)
```
