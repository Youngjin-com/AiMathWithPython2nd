▲리스트6.8：`var()`함수를 사용해 분산을 계산한다

```python
import numpy as np

# 분산을 취하는 데이터
x_1 = np.array([55, 45, 60, 40]) 
x_2 = np.array([51, 49, 52, 48]) 

# 분산의 계산
print(np.var(x_1))
print(np.var(x_2))
```

▲리스트6.9：`std()`함수를 사용해 표준편차를 계산한다

```python
import numpy as np

# 표준편차를 취하는 데이터
x_1 = np.array([55, 45, 60, 40]) 
x_2 = np.array([51, 49, 52, 48]) 

# 표준편차의 계산
print(np.std(x_1))
print(np.std(x_2))
```
