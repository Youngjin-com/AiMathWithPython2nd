▲리스트6.22：`corrcoef()`함수를 사용해서 상관계수를 계산한다

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
 
x = np.array([50, 70, 40, 60, 80])  # 수학 점수
y = np.array([60, 80, 50, 50, 70])  # 영어 점수

print("--- corrcoef() 함수를 사용 ---")
print(np.corrcoef(x, y))  # 상관계수

print()

print("--- 공분산과 표준편차로부터 구한다 ---")
cov_xy = np.average((x-np.average(x))*(y-np.average(y)))  # 공분산
print(cov_xy/(np.std(x)*np.std(y)))  # （식 1）

plt.scatter(x, y)

plt.xlabel("x", size=14)
plt.ylabel("y", size=14)
plt.grid()

plt.show()
```
