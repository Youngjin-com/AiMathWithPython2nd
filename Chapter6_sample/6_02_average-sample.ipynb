▲리스트6.4 : `average()`함수를 사용해 평균값을 계산한다

```python
import numpy as np

x = np.array([55, 45, 60, 40])  # 평균을 취하는 데이터 

print(np.average(x))
```

▲리스트6.5：`sum()` 함수를 사용해 기대값을 계산한다

```python
import numpy as np

p = np.array([0.8, 0.15, 0.05])  # 확률
x = np.array([100, 500, 1000])  # 값

print(np.sum(p*x))  # 기대값
```
